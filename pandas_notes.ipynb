```python
import pandas as pd
#Creation of Dataframe
df=pd.read_csv('monaco_2025_dirty_laps.csv')
print(df.head())
print(df.tail())
print(df.describe())
print(df.info())
print(df.index)
```

Accessing of Data can be done using head tail and etc

```python
df.sample(10)

#LOC AND ILOC FUNCTIONS
print(df.loc[[0,1,5,7]])   #a LIST OF ROWS
print(df.loc[0:3])         #A slice of rows
print(df.loc[0:20,' lap_time_seconds'])  #Range of rows with a specific column
#df.loc[0:[list of columns]] gives a range of columns for a specific rows
df.iloc[0:20,:]   #Same shi but like with indexes on both sides
```

#To convert the index row into the specific column


```python
df.index=df[' lap_number ']
#Now instead of using the index values for rows u can use specific row names
```

#   u    can sort values of a column based on parameters


```python
df1=df
df1.sort_values(' lap_time_seconds', ascending=False)
df1
```

#  A case where u might need to iterate through rows


```python
for index,rows in df.iterrows():
    print(rows[' lap_number '])
```
